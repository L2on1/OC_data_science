## Filtrez les données du data frame

### *.iloc*

`mon_dataframe.iloc[indice_ligne, indice_colonne]`

```python
clients.iloc[0,2]
```

```python
# sélectionner le nom des 10 premiers clients :
clients.iloc[:10, 2]
```

```python
# sélectionner le genre des 10 premiers clients :
clients.iloc[:10, 3]
```

```python
# sélectionner l’e-mail et le genre des 10 derniers clients :
clients.iloc[-10:, [1, 3]]
```

### *.loc*
`mon_dataframe.loc[ condition sur les lignes, colonne(s) ]`

```python
clients.loc[clients["genre"]=="F", :]
```

```python
clients.loc[clients["genre"]=="F", "email"]
```

```python
clients.loc[clients["genre"]=="F", ["email", "genre"]]
```

```python
clients.loc[(clients["genre"]=="F") & (clients["identifiant"]%2==0), ["email", "genre"]]
```

```python
clients.loc[(clients["genre"]=="F") | (clients["identifiant"]%2==0), ["email", "genre"]]
```

```python
clients.sort_values("nom").head()
```

```python
display(clients.loc[0:10, :])
display(clients.iloc[0:10, :])
```

```python
# création d’un data frame trié pour ‘désordonner’ les Index
df_temp = clients.sort_values('nom')
display(df_temp)

# reset_index :
display(df_temp.reset_index())

# reset_index sans garder les anciens Index :
display(df_temp.reset_index(drop=True))
```

On peut modifier des valeurs apres avoir selectionne lignes et colonnes avec *.iloc* ou *.loc*

```python
clients.loc[clients["genre"]=="F", "genre"] = "Femme"
display(clients.loc[clients["genre"]!="M", "genre"])
clients = pd.read_csv('OC_panda_lesson/clients.csv')
```

```python
clients.iloc[:3, 3] = "genre inconnue"
display(clients.head())
```

```python
clients = pd.read_csv('OC_panda_lesson/clients.csv')
display(clients)
```

```python
mask = clients["genre"]=="F"
clients.loc[mask, "identifiant"] += 10e3
display(clients[clients["genre"]!="M"])
clients = pd.read_csv('OC_panda_lesson/clients.csv')
```

## Agrégez des données avec ***Pandas***

```python
prets = pd.read_csv('https://raw.githubusercontent.com/benjaminmrl/data-4452741/main/prets.csv')
```

### Agrégez plusieurs lignes : ***`.group_by`***

Ressemble à du SQL

```python
display(prets.groupby('ville').sum())
```

```python
prets.groupby(['ville', 'type']).sum()
```

```python
prets.groupby(['ville', 'type'])['remboursement'].sum()
```

On peut commencer a faire des choses compliqué

```python
prets.groupby('ville').agg({'remboursement': ['sum', 'mean'],'revenu': 'max'})
```

### Agrégez des lignes et des colonnes : ***`.pivot_table`***

La méthode permettant la transformation inverse s’appelle ***`.melt`***

```python
prets.pivot_table(index='ville', columns='type', values='remboursement', aggfunc='sum')
```

```python
data_pivot = prets.pivot_table(index='ville', columns='type', values='remboursement', aggfunc='sum')
data_pivot.reset_index()
```

```python
data_pivot = data_pivot.reset_index()
data_pivot.melt(id_vars="ville", value_vars=["automobile", "immobilier"])
```

## Fusionnez des données avec Pandas

### Fusion de données comme en SQL : ***`.merge`***

```python
data_merge = pd.merge(clients, prets, on='identifiant')
display(data_merge)
```

Il existe differentes façon pour fusionner les données : 
- Fusion interne (equivalent d'une **Intercetion**)
- Fusion externe (equivalent d'une **Union**)
- Fusion à gauche (on garde que les element de droite et on met NAN quand on sait pas la valeur)
- Fusion à droite (idem que la fusion à gauche)

### Concaténez des données avec Pandas : ***`.concat`***

`pd.concat([df1, df2], ignore_index=True)`

```python
# Client 1
clients_1 = pd.read_csv('https://raw.githubusercontent.com/benjaminmrl/data-4452741/main/clients.csv')
display(clients_1.head())

# Client 2
clients_2 = pd.read_csv('https://raw.githubusercontent.com/benjaminmrl/data-4452741/main/clients_suite.csv')
display(clients_2.head())
```

```python
pd.concat([clients_1, clients_2], ignore_index=True)
```
